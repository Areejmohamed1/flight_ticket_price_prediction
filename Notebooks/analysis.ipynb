```python
import pandas as pd
```

```python
df = pd.read_csv("Cleaned_Data_Train.csv")
```

```python
df
```

```python
def change_into_datetime(col):
    df[col]=pd.to_datetime(df[col])
```

```python
for i in ['Dep_Time', 'Arrival_Time']:
    change_into_datetime(i)
```

```python
def extract_hour(data,col):
    data[col+'_hour']=data[col].dt.hour
    
def extract_min(data,col):
    data[col+'_min']=data[col].dt.minute
    

def drop_col(data,col):
    data.drop(col,axis=1,inplace=True)
```

```python
extract_hour(df,'Dep_Time')

#extracting minutes
extract_min(df,'Dep_Time')

#drop the column
drop_col(df,'Dep_Time')
```

```python
#extracting hour
extract_hour(df,'Arrival_Time')

#extracting min
extract_min(df,'Arrival_Time')


#drop the column
drop_col(df,'Arrival_Time')
```

```python
def flight_length(duration):
    if duration < 180:
        return 'Short'
    elif 180 <= duration < 360:
        return 'Medium'
    else:
        return 'Long'
df['Flight_Length'] = df['Duration(Minutes)'].apply(flight_length)
```

```python
column=[column for column in df.columns if df[column].dtype=='object']
column
```

```python
cat = df[column]
```

```python
Airline=pd.get_dummies(cat['Airline'],drop_first=True).astype(int)
Airline.head()
```

```python
source=pd.get_dummies(cat['Source'],drop_first=True).astype(int)
source.head()
```

```python
destination=pd.get_dummies(cat['Destination'],prefix="Dest",drop_first=True).astype(int)
destination.head()
```

```python
degree = {'Economy': 0, 'Business': 1, 'First Class': 2}
df['Dgree'] = df['Dgree'].map(degree)
```

```python
df['Dgree'].head()
```

```python
Flight_Len =pd.get_dummies(cat['Flight_Length'],drop_first=True).astype(int)
Flight_Len.head()
```

```python
columns_to_drop = ['Airline','Source','Destination','Flight_Length']  
df.drop(columns=columns_to_drop, inplace=True)
```

```python
final_df = pd.concat([Airline,source,destination,Flight_Len,df],axis=1)
final_df.head()
```

```python
print(final_df.isnull().sum())
```

```python
final_df.dropna(subset=['Total_Stops'], inplace=True)
```

```python
final_df.to_csv("final.csv", index=False)
```
